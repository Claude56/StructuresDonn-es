<h3 style="color: SeaGreen" class="fa fa-pencil" >  Les listes chaines: prsentation</h3> 

Il s'agit d'une structure linaire d'lments, qui permet pour des langages comme le C, d'utiliser dynamiquement la mmoire  contrairement aux tableaux du C qui sont du type statique, c'est  dire pas trs efficace lorsqu'il faut les modifier: aggrandir par insertion d'lments ou supprimer des lments et donc rduire l'occupation en mmoire. 
Dans un tableau o tous les lments sont du mme type, l'occupation mmoire de chaque lment est la mme: *taille*. Il suffit alors de stocker en mmoire l'adresse du premier lment: *adresse*. Pour accder  l'lment i, il suffit de calculer son adresse en mmoire: *adresse + i . taille*. Tous les lments sont accessibles avec un cot constant: le temps de calcul de l'adresse + le temps d'accs  cette adresse. 
La taille du tableau est rserve  la cration, donc fixe (statique). Un aggrandissement n'est possible qu' la condition que la mmoire derrire soit libre. Toute insertion d'un lment *i* ncessite de dplacer tous les lments *i+1*. Il n'est pas possible ou difficile de remplacer un lment par un lment d'un autre type (de taille diffrente).  
Solution: stocker en mme temps que la valeur de l'lment, l'adresse de l'lment suivant, c'est la liste chaine. Il n'est plus ncessaire d'utiliser des adresses mmoires contigus. L'accs  un lment *i* se fait en passant par les lments prcdents.

<h3 style="color: navy" >Structure d'une liste chaine:</h3>

Il s'agit de crer une liste o chaque lment est li  ses voisins ( la faon d'une chane):

Cette structure ne peut pas tre directement ralise avec l'objet *list* de Python. La POO nous permet de dfinir cette structure. Pour cela il faut tout d'abord dfinir l'lment de base: le maillon ou cellule.

```python
class Cellule:
    ''''Description de la cellule (ou maillon) d'une liste chane'''
    def __init__(self, v, s):
        self.valeur = v  # La valeur de la cellule
        self.suivant = s # La rfrence de la cellule suivante
```

```python
# Dclaration de la liste:
lstChaine = Cellule(1, Cellule (2, Cellule (3, None))) # None indique la fin de la liste
```

```python
print(lstChaine.valeur)
print(lstChaine.suivant.valeur)
print(lstChaine.suivant.suivant.valeur)
```
